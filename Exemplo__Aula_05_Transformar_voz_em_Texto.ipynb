# Whisper:
O Whisper é um sistema de reconhecimento automático de fala (ASR) treinado em 680.000 horas de dados multilíngues e multitarefa supervisionados coletados da web. Mostramos que o uso de um conjunto de dados tão grande e diversificado leva a uma maior robustez para sotaques, ruído de fundo e linguagem técnica. Além disso, permite a transcrição em vários idiomas, bem como a tradução desses idiomas para o inglês. Somos modelos de código aberto e código de inferência para servir como base para a construção de aplicativos úteis e para pesquisas adicionais sobre processamento de fala robusto.






image.png







image.png



 o Whisper foi treinado em um conjunto de dados grande e diversificado e não foi ajustado para nenhum específico, ele não supera os modelos especializados no desempenho do LibriSpeech, uma referência competitiva famosa em reconhecimento de fala. No entanto, quando medimos o desempenho zero-shot do Whisper em vários conjuntos de dados diversos, descobrimos que ele é muito mais robusto e comete 50% menos erros do que esses modelos.

Cerca de um terço do conjunto de dados de áudio do Whisper não está em inglês e, alternadamente, recebe a tarefa de transcrever no idioma original ou traduzir para o inglês. Achamos que essa abordagem é particularmente eficaz no aprendizado da tradução de fala para texto e supera o SOTA supervisionado na tradução zero do CoVoST2 para o inglês.

```python
import whisper

model = whisper.load_model("small")
result = model.transcribe("caminho do áudio", language="pt")

print(result["text"])
```
